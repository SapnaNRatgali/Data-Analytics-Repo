## *Loan Data Column Overview:*

1. **id**: A unique identifier for each loan to track and manage it.  
2. **address_state**: The state where the borrower resides; used to assess regional risks and trends.  
3. **emp_length**: How long the borrower has been employed, indicating job stability.  
4. **emp_title**: The borrower's job title or occupation, helping to assess income sources.  
5. **grade**: A risk classification based on the borrower’s creditworthiness; higher grade = lower risk.  
6. **sub_grade**: A more detailed risk classification within the Grade, providing further risk differentiation.  
7. **home_ownership**: Indicates whether the borrower owns a home; shows financial stability.  
8. **issue_date**: The date the loan was issued; important for tracking loan aging and interest calculations.  
9. **last_credit_pull_date**: The last date the borrower’s credit report was checked, tracking creditworthiness.  
10. **last_payment_date**: The date the last loan payment was made; used to assess payment behavior.  
11. **loan_status**: The current status of the loan (e.g., paid, active, or defaulted).  
12. **next_payment_date**: The estimated date of the next payment, used for forecasting cash flow.  
13. **purpose**: The reason the borrower took the loan (e.g., debt consolidation, education).  
14. **term**: The duration of the loan, typically in months.  
15. **verification_status**: Indicates if the borrower's financial details were verified for accuracy.  
16. **annual_income**: The borrower’s total yearly income, used to assess repayment ability.  
17. **dti (Debt-to-Income Ratio)**: The ratio of the borrower’s debt payments to their income, used to assess debt burden.  
18. **installment**: The fixed monthly payment amount for loan repayment (principal + interest).  
19. **interest_rate**: The percentage cost of the loan, determining the borrowing cost.  
20. **loan_amount**: The total amount of money borrowed.  
21. **member_id**: A unique identifier for the borrower, helping track their entire relationship with the bank.  
22. **total_acc**: The total number of credit accounts the borrower has, indicating their credit activity and history.  
23. **total_payment**: The total amount paid by the borrower towards the loan, useful for assessing repayment progress and any missed payments.

# *Data Visualization*

## *Charts*

### *1) Bar chart*

```python
# Count the occurrences of each loan purpose
loan_purpose_counts = bl_df['purpose'].value_counts()

# Prepare data for the bar chart
loan_purposes = loan_purpose_counts.index.tolist()  # List of loan purposes
counts = loan_purpose_counts.values.tolist()  # Corresponding counts

# Create a bar chart for loan purpose distribution
plt.figure(figsize=(8, 5))  # Set the figure size

# Plot the bar chart
plt.bar(loan_purposes, counts, color="Lavender")

# Add title and labels
plt.title('Loan Purpose Distribution', fontsize=14)
plt.xlabel('Loan Purpose', fontsize=12)
plt.ylabel('Number of Loans', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=55)

# Add grid lines for better readability
plt.grid(axis='y', alpha=0.7)

# Show the plot
plt.show()

```

#### *Insight:*
##### *The bar chart displays the distribution of loan purposes, showing the number of loans for each category.*
##### *It helps to easily identify which loan purposes are most common in the dataset.*



### *2) Barh chart*

```python
emp_title_counts = bl_df['emp_title'].value_counts().head(10)  # Top 10 titles for clarity
plt.figure(figsize=(8, 4))
plt.barh(emp_title_counts.index, emp_title_counts.values, color="pink") # index is the job title, values are the counts .
plt.xlabel("Number of Borrowers (Count)")  # Adding units to the x-axis
plt.ylabel("Employment Title")  # The y-axis doesn't need units in this case, as it's categorical
plt.title("Top 10 Employment Titles of Borrowers")
plt.show()
```

#### *Insight:*
##### *The chart shows the most common jobs of loan applicants, helping banks understand the types of jobs borrowers have, which can assist in evaluating their ability to repay loans.*


### *3) Stacked Bar chart*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count loans by employment length and loan status
loan_count_by_emp_length = bl_df.groupby(['emp_length', 'loan_status']).size().unstack()

# Create a bar plot for employment length vs loan status
plt.figure(figsize=(12, 6))
loan_count_by_emp_length.plot(kind='bar', stacked=True, color=['brown', 'yellow', 'green'])

# Labeling the axes and the plot
plt.xlabel('Employment Length (Years)')  # Adding units to the x-axis
plt.ylabel('Number of Loans (Count)')    # Adding units to the y-axis
plt.title('Loan Status Distribution by Employment Length')
plt.xticks(rotation=45)
plt.legend(title='Loan Status')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

#### *Insight*
##### *The chart shows how loans are repaid or defaulted based on how long borrowers have been employed.*
##### *It helps the bank understand if people with more work experience are more likely to repay their loans.*
- **Brown:** Loan is Fully Paid.
- **Yellow:** Loan is Current (ongoing)
- **Green:** Loan is Charged Off (defaulted).

### *5) Line chart*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'issue_date' is in datetime format with day first
bl_df['issue_date'] = pd.to_datetime(bl_df['issue_date'], dayfirst=True)

# Group loans by month-end and loan status, then count them
loan_status_over_time = bl_df.groupby([pd.Grouper(key='issue_date', freq='ME'), 'loan_status']).size().unstack()

# Create a plot
loan_status_over_time.plot(figsize=(10, 6), marker='o', linestyle='-')

# Label axes and add title with units
plt.xlabel('Issue Date (Month)', fontsize=12)  # x-axis: Issue Date in months
plt.ylabel('Number of Loans (Count)', fontsize=12)  # y-axis: Number of Loans issued
plt.title('Loan Status Over Time (Month-wise)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)

# Display the plot
plt.show()
```

#### *Insight*
##### *The line chart displays the monthly distribution of loan statuses (Fully Paid, Current, and Charged Off), helping to track trends over time.* 
##### *It provides insights into how loan repayment behavior has evolved month by month.*

### *6) Histogram chart*

```python
# Plot the histogram and store patches (bars)
n, bins, patches = plt.hist(bl_df['loan_amount'], bins=20, edgecolor='black')

# Apply different colors to each bar
for i, patch in enumerate(patches):
    color = plt.cm.viridis(i / len(patches))  # Get color for the current bar based on its index
    patch.set_facecolor(color)                  # Set the color for each bar

# Add labels and title
plt.xlabel('Loan Amount (USD)')                 # Adding unit for the x-axis (currency)
plt.ylabel('Frequency (Number of Loans)')       # Adding unit for the y-axis (count of loans)
plt.title('Loan Amount Distribution')           # Title of the histogram

# Show the plot
plt.show()                                      # Display the histogram
```

#### *Insight*
##### *The histogram shows the distribution of loan amounts given by the bank.*
##### *It helps identify the most common loan amounts, giving the bank an idea of the usual loan amounts requested by customers.*

### *7) Grouped Bar chart*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by 'grade' and 'loan_status' and count the number of occurrences
loan_status_by_grade = bl_df.groupby(['grade', 'loan_status']).size().unstack()

# Plot as a simple grouped bar chart
loan_status_by_grade.plot(kind='bar', figsize=(10, 6))

# Add labels and title with units
plt.xlabel('Loan Grade (Categorical)')           # X-axis label with unit (categorical)
plt.ylabel('Number of Loans (Count)')            # Y-axis label with unit (count of loans)
plt.title('Loan Status Distribution by Grade')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

### *8) Pie chart*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'bl_df' is the DataFrame that contains your loan data

# Count the occurrences of each grade
grade_counts = bl_df['grade'].value_counts()

# Count the occurrences of each loan status
loan_status_counts = bl_df['loan_status'].value_counts()

# Create a figure with a specific size
plt.figure(figsize=(8, 5))

# Plot for loan status distribution (left side)
plt.subplot(1, 2, 1)  # (rows, columns, position)
loan_status_counts.plot.pie(
    explode=[0.1] * len(loan_status_counts),  # Add slight separation for each slice
    autopct='%1.1f%%',  # Display percentage with one decimal point
    shadow=True,        # Add shadow effect for a 3D look
    colors=['lightcoral', 'lightgreen']  # Customize colors for slices
)
plt.title('Loan Status Distribution')

#------------------------------------------------------------------------------------------------

# Plot for loan distribution by grade (right side)
plt.subplot(1, 2, 2)  # (rows, columns, position)
grade_counts.plot.pie(
    explode=[0.1] * len(grade_counts),  # Add slight separation for each slice
    autopct='%1.1f%%',  # Display percentage with one decimal point
    shadow=True,        # Add shadow effect for a 3D look
    colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue']  # Customize colors for slices
)
plt.title('Distribution of Loans by Grade')

# Adjust layout to prevent overlap of titles/labels
plt.tight_layout()

# Show the plot
plt.show()
```

#### *Insight*
##### *The left chart shows the loan status distribution (approved/declined or other status).*
##### *The right chart shows the distribution of loans by grade (A, B, C, D, etc.).*
